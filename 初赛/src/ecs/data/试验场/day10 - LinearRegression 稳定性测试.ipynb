```python
from linalg.common import dim,shape,reshape
from linalg.matrix import matrix_inverse,matrix_transpose,matrix_mutmul,zeros,matrix_minus,identity_matrix,matrix_multiply_factor,col,matrix_copy
from linalg.vector import mean,squrt,root,minus,dot
import copy
```

```python
#! -*- coding: utf-8 -*-
import numpy as np

class Lasso():
    def __init__(self, alpha=1.0, n_iter=1000):
        self.alpha = alpha 
        self.n_iter = n_iter 
        self.W = None

    def _soft_thresholding_operator(self, x, lambda_):
        if x > 0 and lambda_ < abs(x):
            print(1)
            return x - lambda_
        elif x < 0 and lambda_ < abs(x):
            print(2)
            return x + lambda_
        else:
            print(3)
            return 0

    def fit(self, X, y):
        X_t = matrix_transpose(X)
        M,N = shape(X)
        self.W = zeros((N,1),dtype=float)
        for _ in range(self.n_iter):
            for j in range(N):
#                 t_W = matrix_copy(self.W)
#                 t_W[j][0] = 0.0
                y_ = reshape(matrix_mutmul(X,self.W),(-1,))
                arg1 = dot(X_t[j], y_)
                arg2 = self.alpha * M
                self.W[j][0] = self._soft_thresholding_operator(arg1, arg2)/sum(squrt(X_t[j]))
                
    def predict(self, X):
        assert(self.W!=None)
        return matrix_mutmul(X,self.W)



```

```python
clf = Lasso()
clf.fit(_X_train,_Y_train)
p = clf.predict(_X_test)
np.power(np.sum((np.array(p) - Y_test)**2),0.5)
```
